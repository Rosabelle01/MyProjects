```python
# import the pandas library
import pandas as pd
```

```python
# read in the csv file as a pandas dataframe
cases = pd.read_csv('Case_Information.csv')
```

```python
# explore the data
cases.info()
```

```python
cases.shape
```

```python
#display the dataframe
cases
```

```python
#remove the duplicate rows
cases = cases.drop_duplicates()
cases
```

```python
#remove the unuseful columns
cases = cases.drop(columns = 'date_announced_as_removed' )
cases
```

```python
cases = cases.drop(columns = 'muni_city' )
cases
```

```python
cases = cases.drop(columns = 'health_status' )
cases
```

```python
cases = cases.drop(columns = 'date_recovered' )
cases
```

```python
cases = cases.drop(columns = 'date_of_death')
cases
```

```python
cases = cases.drop(columns = 'home_quarantined')
cases
```

```python
cases = cases.drop(columns = 'date_of_onset_of_symptoms')
cases
```

```python
cases = cases.drop(columns = 'pregnant')
cases
```

```python
cases = cases.drop(columns = 'region')
cases
```

```python
#remove all the NaN values
cases.fillna('')
```

```python
#reset index
cases.reset_index(drop=True)
cases
```
